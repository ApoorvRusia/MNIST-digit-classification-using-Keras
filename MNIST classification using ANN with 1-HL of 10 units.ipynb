```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
datasettrain = pd.read_csv('mnist_train.csv')
datasettest = pd.read_csv('mnist_test.csv')
X_train = datasettrain.iloc[:, 1:].values
y_train = datasettrain.iloc[:, 0].values
X_test = datasettest.iloc[:, 1:].values
y_test = datasettest.iloc[:, 0].values
m = np.identity(10)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
y_train = [m[i,:] for i in y_train]
y_train = pd.DataFrame(y_train)
y_train = y_train.iloc[:,:].values
```

```python
# Part 2 - Now let's make the ANN!

# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
from keras.layers import Dense

# Initialising the ANN
classifier = Sequential()

# Adding the input layer and the first hidden layer
classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu', input_dim = 784))

# Adding the second hidden layer
#classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'relu'))

# Adding the output layer
classifier.add(Dense(units = 10, kernel_initializer = 'uniform', activation = 'softmax'))

# Compiling the ANN
classifier.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])

# Fitting the ANN to the Training set
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
```

```python
y_pred = np.argmax(y_pred, axis=1)
```

```python
from sklearn.metrics import confusion_matrix
conMat = confusion_matrix(y_test, y_pred)
```

```python
#Calculating accuracy
totalCorrect = 0

for i in range(conMat.shape[0]):
    for j in range(conMat.shape[1]):
        if (i == j):
            totalCorrect += conMat[i,j]
print('The accuracy of the model is {}'.format(100*totalCorrect/y_test.size))
```
